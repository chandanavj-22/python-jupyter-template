### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

def plot_subplots(x, y, categories, values, labels, sizes):
    try:
        # Validate inputs
        if not all(isinstance(lst, list) for lst in [x, y, categories, values, labels, sizes]):
            raise TypeError("All inputs must be lists.")

        if len(x) != len(y):
            raise ValueError("x and y must have the same length.")
        
        if len(categories) != len(values):
            raise ValueError("Categories and values must have the same length.")
        
        if len(labels) != len(sizes):
            raise ValueError("Labels and sizes must have the same length.")

        if not all(isinstance(i, (int, float)) for i in x + y + values + sizes):
            raise TypeError("Numeric lists must contain only integers or floats.")

        # Create a 2x2 grid of subplots
        fig, axs = plt.subplots(2, 2, figsize=(10, 10))
        fig.suptitle("Different Types of Plots", fontsize=14, fontweight='bold')

        # First subplot: Line plot
        axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
        axs[0, 0].set_title('Line Plot')

        # Second subplot: Bar plot
        axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
        axs[0, 1].set_title('Bar Plot')

        # Third subplot: Scatter plot
        axs[1, 0].scatter(x, y, color='orange')
        for i, (xi, yi) in enumerate(zip(x, y)):  # Add annotations
            axs[1, 0].text(xi, yi, f'({xi},{yi})', fontsize=9, ha='right')
        axs[1, 0].set_title('Scatter Plot')

        # Fourth subplot: Pie chart
        axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightcoral', 'lightgreen'])
        axs[1, 1].set_title('Pie Chart')

        # Adjust layout
        plt.tight_layout(rect=[0, 0, 1, 0.96])

        # Show the plots
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]

plot_subplots(x, y, categories, values, labels, sizes)
```
