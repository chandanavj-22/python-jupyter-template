### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

def plot_histogram(data, bins=5, color='blue'):
    try:
        # Check if data is a list
        if not isinstance(data, list):
            raise TypeError("Data must be a list.")

        # Ensure data list is not empty
        if len(data) == 0:
            raise ValueError("Data list cannot be empty.")

        # Ensure all elements in data are numeric
        if not all(isinstance(i, (int, float)) for i in data):
            raise TypeError("All elements in data must be numeric.")

        # Ensure bins is a positive integer
        if not isinstance(bins, int) or bins <= 0:
            raise ValueError("Bins must be a positive integer.")

        # Create the histogram with a user-defined color
        plt.hist(data, bins=bins, color=color, edgecolor='black')

        # Labels and title
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.title('Histogram Example')

        # Show the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
data = [1, 2, 2, 3, 4, 5, 5, 5, 6, 7]
plot_histogram(data, bins=5, color='green')
```
