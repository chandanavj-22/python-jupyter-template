### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

def plot_histogram(data, bins=5):
    try:
        if not isinstance(data, list):
            raise TypeError("Data must be a list.")
        if len(data) == 0:
            raise ValueError("Data list cannot be empty.")
        if not all(isinstance(i, (int, float)) for i in data):
            raise TypeError("All elements in data must be numeric.")

        # Create a histogram
        plt.hist(data, bins=bins, color='blue', edgecolor='black')

        # Labeling the axes and the plot
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.title('Histogram Example')
        # Show the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example data
data = [1, 2, 2, 3, 4, 5, 5, 5, 6, 7]

plot_histogram(data)
```
