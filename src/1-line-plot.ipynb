### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.

import matplotlib.pyplot as plt

def plot_line(x, y, title="Line Plot"):
    try:
        # Validate inputs
        if not isinstance(x, list) or not isinstance(y, list):
            raise TypeError("Both x and y must be lists.")

        if len(x) != len(y):
            raise ValueError("x and y must have the same length.")

        if not all(isinstance(i, (int, float)) for i in x + y):
            raise TypeError("All elements in x and y must be numeric.")

        # Create a line plot
        plt.plot(x, y, marker='o', linestyle='-', color='b')

        # Labeling the axes and the plot
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title(title)
        plt.grid(True)  # Adding grid for better readability

        # Show the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
plot_line(x, y, title="Custom Line Plot")
```
