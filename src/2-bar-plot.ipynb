### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

def plot_bar_chart(categories, values):
    try:
        if len(categories) != len(values):
            raise ValueError("The number of categories must match the number of values.")
        
        plt.bar(categories, values, color='skyblue')
        plt.xlabel('Categories')
        plt.ylabel('Values')
        plt.title('Bar Plot Example')
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [4, 7, 3, 6]

plot_bar_chart(categories, values)
```
