### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

def plot_bar_chart(categories, values):
    try:
        # Check if inputs are lists
        if not isinstance(categories, list) or not isinstance(values, list):
            raise TypeError("Both categories and values must be lists.")

        # Check if lists have the same length
        if len(categories) != len(values):
            raise ValueError("Categories and values lists must have the same length.")

        # Ensure all values are numeric
        if not all(isinstance(v, (int, float)) for v in values):
            raise TypeError("All elements in values must be numeric.")

        # Define different colors for each bar
        colors = ['red', 'blue', 'green', 'purple', 'orange']
        colors = colors[:len(categories)]  # Adjust to match the number of categories

        # Create bar plot with different colors
        plt.bar(categories, values, color=colors)

        # Labels and title
        plt.xlabel('Categories')
        plt.ylabel('Values')
        plt.title('Bar Plot with Different Colors')

        # Show the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example data
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]

plot_bar_chart(categories, values)
```
