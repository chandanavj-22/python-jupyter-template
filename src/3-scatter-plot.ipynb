### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

def plot_scatter(x, y):
    try:
        # Check if inputs are lists
        if not isinstance(x, list) or not isinstance(y, list):
            raise TypeError("Both x and y must be lists.")
        
        # Check if both lists have the same length
        if len(x) != len(y):
            raise ValueError("x and y must have the same length.")
        
        # Ensure all elements are numeric
        if not all(isinstance(i, (int, float)) for i in x + y):
            raise TypeError("All elements in x and y must be numeric.")
        
        # Create scatter plot
        plt.scatter(x, y, color='blue', label='Data Points')

        # Add annotations for each point
        for i, (xi, yi) in enumerate(zip(x, y)):
            plt.text(xi, yi, f'({xi},{yi})', fontsize=10, ha='right')

        # Labels and title
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Scatter Plot with Annotations')
        plt.legend()

        # Show the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

plot_scatter(x, y)
```
